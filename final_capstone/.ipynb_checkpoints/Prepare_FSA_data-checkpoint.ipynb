```python
fsa_list_bc=fsa_bc['Code'].to_list()
```

```python
def process_fsa(fsa):
    def build_foursquare_endpoint_url(near_str):
        client_id='GI1CV0OGIPJVB2SOW4TL03H1OBYE0KH5BSJTWHW2SFXISANF'
        client_secret='BENPMPAJAD5M3RD0O1N4LHLIQTXPBOJBFXU3M0MSJZXM0YOW'
        v='20201101'
        near=near_str
        intent='browse'
        radius=10000
        query='Restaurant'
        limit=50

        endpoint_url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&v={}&near={}&intent={}&radius={}&query={}&limit={}'.format(client_id,client_secret, v, near,intent,radius,query, limit)

        return endpoint_url

    venues_categories=[]
    url = build_foursquare_endpoint_url(fsa)
    r=requests.get(url)
    print(r.status_code)
    venues_results = r.json()["response"]['venues']
    for v in venues_results:
        try:
            venues_categories.append({'FSA':fsa,'lat':v['location']['lat'], 'lon':v['location']['lng'], 'category':v['categories'][0]['shortName']})
        except IndexError:
            # return venues_categories.append({'FSA':'','lat':'', 'lon':'', 'category':''})
            return None
    
    return venues_categories
```

```python
import pandas as pd

master_venue_categories = []

for a_fsa in fsa_list_bc:
    fsa_df = pd.DataFrame(process_fsa(a_fsa))
    print(len(fsa_df))
    if len(fsa_df)>0:
        master_venue_categories.append(fsa_df)
```

```python
df = pd.concat(master_venue_categories)
```
